```python
from datasets import load_dataset
ds = load_dataset('bunkalab/medium-sample-technology')['train'].shuffle(seed=42)
docs = ds['title']
docs[:10]
```

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer(model_name_or_path='BAAI/bge-small-en-v1.5')
```

```python
from topicviz import TopicViz
tv = TopicViz(n_topics=12, random_state=42)
tv.fit(docs, embedding_model=embedding_model)
```

```python
fig = tv.visualize_topics(
    width=800, height=800,
    colorscale='Blues',
    density=True,
    show_text=True,
    repel_labels=False,
    topic_label_mode='embedding',
)
fig.show()
```

```python
fig_b = tv.visualize_bourdieu(
    docs=docs,  
    embedding_model=embedding_model,
    x_left_words=["This shows high autonomy satisfaction"],
    x_right_words=["This shows low autonomy satisfaction"],
    y_top_words=["This shows high autonomy frustration"],
    y_bottom_words=["This shows low autonomy frustration"],
    radius_size=0.2,
    height=800,
    width=800,
    clustering=False,
    density=False,
    convex_hull=True,
)
```

```python
fig_b.show()
```
